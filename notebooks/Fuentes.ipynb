```python

from pathlib import Path
USER_CSV = Path("../data/quejas-clientes.csv")  # ajusta si usas tu archivo
CSV_PATH = USER_CSV
#Opciones de pandas
pd.set_option("display.max_colwidth", 120)
pd.set_option("display.max_columns", 120)

df = pd.read_csv(CSV_PATH, sep=',', encoding='utf-8')
print("Shape:", df.shape)
df.head(5)
```

```python

# Información general
display(pd.DataFrame({"column": df.columns, "dtype": df.dtypes.astype(str)}))
print("\nValores faltantes por columna:")
display(df.isna().sum().sort_values(ascending=False).to_frame("missing"))
```

```python

# Distribución objetivo si existe
if "Timely response?" in df.columns:
    print("Distribución 'Timely response?':")
    display(df["Timely response?"].astype("string").str.strip().str.lower().value_counts(dropna=False))
```

```python

# Guardar una muestra aleatoria de 1k filas para pruebas rápidas
sample_path = Path("data/samples/quick_sample.csv")
sample_path.parent.mkdir(parents=True, exist_ok=True)
df.sample(n=min(1000, len(df)), random_state=42).to_csv(sample_path, index=False)
print("Muestra guardada en:", sample_path.resolve())
```
